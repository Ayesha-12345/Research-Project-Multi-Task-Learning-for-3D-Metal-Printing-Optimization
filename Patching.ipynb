```python
import os
import cv2

def patch_images(folder_path, patch_size, stride, destination_folder):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Get a list of image files in the folder
    image_files = [file for file in os.listdir(folder_path) if (file.endswith('.png') or file.endswith('.jpg'))]

    # Sort the files alphabetically
    image_files.sort()

    for file in image_files:
        # Read the image
        image_path = os.path.join(folder_path, file)
        image = cv2.imread(image_path)

        # Get image dimensions
        height, width, _ = image.shape

        # Extract patches from the image
        patch_num = 1
        for i in range(0, height - patch_size + 1, stride):
            for j in range(0, width - patch_size + 1, stride):
                patch = image[i:i+patch_size, j:j+patch_size]

                # Save the patch with appropriate name in the destination folder
                patch_name = f"{os.path.splitext(file)[0]}_{i}-{j}.jpg"
                patch_path = os.path.join(destination_folder, patch_name)
                cv2.imwrite(patch_path, patch)
                print(f"Saved patch '{patch_name}'")

patch_images(r"/home/mishra/Documents/Ayesha/Ayesha 2/FInal selected Raw Data/BD Parallel",
             patch_size=800,
             stride=200,
             destination_folder=r"/home/mishra/Documents/Ayesha/Ayesha 2/FInal selected Raw Data/800x800 PATCHED/BD Parallel")
```

```python
import os
from PIL import Image

source_folder = r"/home/mishra/Documents/Ayesha/Ayesha 2/FInal selected Raw Data/800x800 PATCHED/BD Normal"
destination_folder = '/home/mishra/Documents/Ayesha/Ayesha 2/FInal selected Raw Data/800x800 Patched and flipped/BD Normal'

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Iterate through the files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith('.jpg'):
        # Open the image
        image_path = os.path.join(source_folder, filename)
        image = Image.open(image_path)

        # Save the original image
        original_name = os.path.splitext(filename)[0] + '_original.jpg'
        original_path = os.path.join(destination_folder, original_name)
        image.save(original_path)

        # Flip horizontally
        flipped_horizontal = image.transpose(Image.FLIP_LEFT_RIGHT)
        flipped_horizontal_name = os.path.splitext(filename)[0] + '_horizontal.jpg'
        flipped_horizontal_path = os.path.join(destination_folder, flipped_horizontal_name)
        flipped_horizontal.save(flipped_horizontal_path)

        # Flip vertically
        flipped_vertical = image.transpose(Image.FLIP_TOP_BOTTOM)
        flipped_vertical_name = os.path.splitext(filename)[0] + '_vertical.jpg'
        flipped_vertical_path = os.path.join(destination_folder, flipped_vertical_name)
        flipped_vertical.save(flipped_vertical_path)

        # Flip both horizontally and vertically
        flipped_both = image.transpose(Image.ROTATE_180)
        flipped_both_name = os.path.splitext(filename)[0] + '_both.jpg'
        flipped_both_path = os.path.join(destination_folder, flipped_both_name)
        flipped_both.save(flipped_both_path)

print("Data augmentation complete.")
```

```python
import os
from openpyxl import Workbook

folder_path = '/home/mishra/Documents/Ayesha/Ayesha 2/FInal selected Raw Data/800x800 Patched and flipped/BD Normal'


# Create a new workbook
workbook = Workbook()
sheet = workbook.active

# Get the list of files in the folder
file_list = os.listdir(folder_path)

# Iterate through the files in the folder without sorting
for filename in file_list:
    if filename.endswith('.jpg'):
        # Get the complete file path
        file_path = os.path.join(folder_path, filename)

        # Add the file path to the Excel sheet
        sheet.append([file_path])

# Save the workbook
excel_file_path = os.path.join(folder_path, 'Normal.xlsx')
workbook.save(excel_file_path)

print("Excel sheet created with image file paths.")
```
